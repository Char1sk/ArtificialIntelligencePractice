```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_result(path):
    with open(path) as f:
        lines = f.readlines()
        trainmiou = []
        testmiou = []
        for l in lines:
            if 'Train' in l:
                # i = float(l[31:37])
                idx = l.find('mIOU') + 5
                i = float(l[idx:idx+6])
                trainmiou.append(i)
            if 'Test' in l:
                # i = float(l[30:36])
                idx = l.find('mIOU') + 5
                i = float(l[idx:idx+6])
                testmiou.append(i)
    # print(trainmiou)
    # print(testmiou)
    print(max(trainmiou))
    print(max(testmiou))
    return trainmiou, testmiou, max(trainmiou), max(testmiou)
```

```python
u11 = get_result('./exp01.txt')
u12 = get_result('./unet2_lr1.txt')
plt.plot(u11[1])
plt.plot(u12[1])
plt.legend(['UNet1', 'UNet2'])
plt.show()
```

```python
with open('./DeepLab-xception-2.txt') as f:
    lines = f.readlines()
    trainloss = []
    testloss = []
    trainmiou = []
    testmiou = []
    for l in lines:
        if 'Train' in l:
            # i = float(l[31:37])
            idx1 = l.find('Loss') + 5
            idx2 = l.find(',')
            i = float(l[idx1:idx2])
            trainloss.append(i)
            
            idx3 = l.find('mIOU') + 5
            i = float(l[idx3:idx3+6])
            trainmiou.append(i)
        if 'Test' in l:
            # i = float(l[30:36])
            idx1 = l.find('Loss') + 5
            idx2 = l.find(',')
            i = float(l[idx1:idx2])
            testloss.append(i)
            
            idx = l.find('mIOU') + 5
            i = float(l[idx:idx+6])
            testmiou.append(i)
# print(trainmiou)
# print(testmiou)
print(max(trainloss))
print(max(testloss))
print(max(trainmiou))
print(max(testmiou))
```
